# Visualize the distribution of content across release years.

```python
df
```

```python
content_year = df.groupby("release_year")["release_year"].value_counts().sort_values(ascending = False).reset_index()
content_year.head(10)
```

```python
plt.pie(content_year['count'].head(10), autopct='%.1f%%',labels=content_year.release_year.head(10))
plt.title("Content Released over the years")
plt.show()
```

## Insights from the Data Visualization

From the above graph, we can observe a significant increase in movie releases during the years 2017, 2018, 2019, and 2020. However, in 2021, there was a noticeable decline in the number of movie releases. This decrease can be attributed to the impact of the COVID-19 pandemic on the cinema industry.

### Factors Contributing to the Decline:

1. **Cinema Closures:** Movie theaters were closed in many regions due to the pandemic, leading to a reduction in movie releases. The closure or limited operation of theaters affected studios' willingness to release big-budget films, as they heavily rely on box office revenue.

2. **COVID-19 Pandemic:** The pandemic disrupted film production schedules globally. Lockdowns, travel restrictions, and safety protocols caused delays in shooting, post-production, and distribution. Consequently, studios had to adjust their release calendars accordingly.

3. **Production Challenges:** Filming during the pandemic required adherence to strict safety protocols, including testing, social distancing, and limited crew sizes. These challenges slowed down production timelines, further contributing to the decline in movie releases.

Overall, the data suggests that the COVID-19 pandemic had a significant impact on the cinema industry, resulting in fewer movie releases in 2021 compared to previous years.


# Analyze the distribution of content ratings.

```python
ratings = df.groupby("rating")["rating"].value_counts().sort_values(ascending = False).reset_index()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(ratings,x = "rating",y = "count")
plt.title("Rating based Content")
```

```python
plt.figure(figsize=(12,6))
plt.pie(ratings["count"].head(5),labels=ratings.rating.head(5),  autopct='%.1f%%')
plt.title("Rating Based Content For Top 5")
plt.show()
```

# Explore the length of movies or episodes and identify any trends.

```python
plt.figure(figsize=(12,6))
sns.barplot(data = time, x = "duration", y = "count")
plt.title("Duration Analysis For Top 10")
plt.show()
```

# Further explore the distribution of content across different countries and regions.

```python
dis_country = df[df["country"] == "United States"][["type", "rating"]].value_counts().reset_index()
dis_country
```

```python
print("Distribution of Movies and TV Shows from the United States by Rating")
sns.catplot(kind = "bar", data = dis_country, x = "count", y = "rating", col = "type")
plt.show()
```

```python
plt.pie(dis_country[dis_country["type"] == "Movie"]["count"].head(5),autopct='%.1f%%', labels = dis_country[dis_country["type"] == "Movie"]["rating"].head(5))
plt.title("Top 5 Movie Rating in United States")
plt.show()
```

```python
plt.pie(dis_country[dis_country["type"] == "TV Show"]["count"],autopct='%.1f%%', labels = dis_country[dis_country["type"] == "TV Show"]["rating"])
plt.title("Top TV Shows Rating in United States")
plt.show()
```
